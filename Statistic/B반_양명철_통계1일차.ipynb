```python
import numpy as np
from scipy import stats
import scipy.stats
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.proportion import proportions_ztest
```

```python
# myscars 데이터 가져오기(데이터의 경로 확인: mycars.csv)
ds_mycars = pd.read_csv("mycars.csv", engine = "python")
ds_mycars
```

```python
df = ds_mycars[['manufacturer', 'model', 'year', 'cylinder', 'automatic', 'driving', 'fuel', 'class', 'mpg']]
df
```

```python
count_manu = df['manufacturer'].value_counts().sort_index()
cumcnt_manu = np.cumsum(count_manu)
percent_manu = count_manu / sum(count_manu) * 100
cumpct_manu = np.cumsum(count_manu)
```

```python
count_model = df['model'].value_counts().sort_index()
cumcnt_model = np.cumsum(count_model)
percent_model = count_model / sum(count_model) * 100
cumpct_model = np.cumsum(percent_model)
```

```python
count_year = df['year'].value_counts().sort_index()
cumcnt_year = np.cumsum(count_year)
percent_year = count_year / sum(count_year) * 100
cumpct_year = np.cumsum(percent_year)
```

```python
count_cylinder = df['cylinder'].value_counts().sort_index()
cumcnt_cylinder = np.cumsum(count_cylinder)
percent_cylinder = count_cylinder / sum(count_cylinder) * 100
cumpct_cylinder = np.cumsum(count_cylinder)
```

```python
count_auto = df['automatic'].value_counts().sort_index()
cumcnt_auto = np.cumsum(count_auto)
percent_auto = count_auto / sum(count_auto) * 100
cumpct_auto = np.cumsum(percent_auto)
```

```python
count_dri = df['driving'].value_counts().sort_index()
cumcnt_dri = np.cumsum(count_dri)
percent_dri = count_dri / sum(count_dri) * 100
cumpct_dri = np.cumsum(percent_dri)
```

```python
count_fuel = df['fuel'].value_counts().sort_index()
cumcnt_fuel = np.cumsum(count_fuel)
percent_fuel = count_fuel / sum(count_fuel) * 100
cumpct_fuel = np.cumsum(percent_fuel)
```

```python
count_class = df['class'].value_counts().sort_index()
cumcnt_class = np.cumsum(count_class)
percent_class = count_class / sum(count_class) * 100
cumpct_class = np.cumsum(percent_class)
```
